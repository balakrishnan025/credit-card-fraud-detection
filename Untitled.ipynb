# Credit Card Fraud Detection (Beginner Project)

In this project, we use the Kaggle credit card fraud dataset (284,807 transactions, 492 frauds) to build simple machine learning models that classify each transaction as normal (0) or fraud (1).


```python
# Import pandas library
import pandas as pd

# Read the CSV file into a DataFrame
# Make sure 'creditcard.csv' is in the same folder as this notebook
data = pd.read_csv("creditcard.csv")

# Show the first 5 rows
data.head()

```

We load the CSV file into a pandas DataFrame called `data`.  
Then we look at the first few rows and some basic information to understand the size and columns of the dataset.


```python
print("Shape (rows, columns):", data.shape)
```

```python
data.describe()
```

```python
print("Shape (rows, columns):", data.shape)

# Column types and non‑null counts
data.info()
```

```python
missing_values = data.isnull().sum()
print(missing_values)
```

```python
class_counts = data['Class'].value_counts()
print(class_counts)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(6, 4))

sns.barplot(x=class_counts.index, y=class_counts.values)

plt.title("Class distribution (0 = normal, 1 = fraud)")
plt.xlabel("Class")
plt.ylabel("Number of transactions")

plt.show()
```

Here we check if there are any missing values and see how many normal vs fraud transactions we have.  
The plot shows that fraud cases are much fewer than normal ones, so the dataset is highly imbalanced.


```python
fraud = data[data['Class'] == 1]

# All rows where Class == 0 (normal)
normal = data[data['Class'] == 0]

print("Number of fraud transactions:", len(fraud))
print("Number of normal transactions:", len(normal))
```

```python
normal_sample = normal.sample(n=len(fraud), random_state=42)

print("Sampled normal transactions:", len(normal_sample))
```

```python
balanced_data = pd.concat([fraud, normal_sample], axis=0)

# Shuffle the rows so fraud and normal are mixed together
balanced_data = balanced_data.sample(frac=1, random_state=42).reset_index(drop=True)

# Check the new class distribution
print("Balanced class counts:")
print(balanced_data['Class'].value_counts())
```

Because there are so many more normal transactions than fraud ones, we create a smaller, balanced dataset. 
    
We keep all fraud rows and randomly pick the same number of normal rows, then shuffle them together.


```python
# Features: all columns except 'Class'
X = balanced_data.drop('Class', axis=1)

# Label: the 'Class' column
y = balanced_data['Class']

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
from sklearn.model_selection import train_test_split

# Split into train and test sets
# test_size=0.2 -> 20% test, 80% train
# stratify=y keeps the class balance in both sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)

print("\nTrain set class counts:")
print(y_train.value_counts())

print("\nTest set class counts:")
print(y_test.value_counts())
```

We split the balanced dataset into features `X` (all columns except `Class`) and label `y` (`Class` column).  
Then we make a training set (80%) and test set (20%) using `train_test_split`, keeping the class balance.


```python
from sklearn.linear_model import LogisticRegression

# Create the Logistic Regression model
# max_iter is increased a bit so the algorithm has enough time to converge
log_reg_model = LogisticRegression(
    max_iter=1000,      # maximum number of training iterations
    solver='liblinear'  # a good solver for small binary problems
)

# Train (fit) the model on the training data
log_reg_model.fit(X_train, y_train)

print("Logistic Regression model trained successfully!")
```

```python
# Use the trained model to predict labels for the test set
y_pred = log_reg_model.predict(X_test)

# Look at the first 10 predictions and true labels
print("Predicted labels:", y_pred[:10])
print("True labels     :", y_test.values[:10])
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)

# Compute basic metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)  # by default: positive class is '1'
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Logistic Regression performance on test set:")
print("Accuracy :", accuracy)
print("Precision:", precision)
print("Recall   :", recall)
print("F1-score :", f1)

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(cm)

# Optional: more detailed report per class
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

# Create the Random Forest model
rf_model = RandomForestClassifier(
    n_estimators=100,   # number of trees in the forest
    random_state=42,    # for reproducibility
    n_jobs=-1           # use all CPU cores to speed up training
)

# Train (fit) the Random Forest model
rf_model.fit(X_train, y_train)

print("Random Forest model trained successfully!")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

# Predict on the test set
y_pred_rf = rf_model.predict(X_test)

# Compute metrics
accuracy_rf = accuracy_score(y_test, y_pred_rf)
precision_rf = precision_score(y_test, y_pred_rf)
recall_rf = recall_score(y_test, y_pred_rf)
f1_rf = f1_score(y_test, y_pred_rf)

print("Random Forest performance on test set:")
print("Accuracy :", accuracy_rf)
print("Precision:", precision_rf)
print("Recall   :", recall_rf)
print("F1-score :", f1_rf)

# Confusion matrix for Random Forest
cm_rf = confusion_matrix(y_test, y_pred_rf)
print("\nRandom Forest Confusion Matrix:")
print(cm_rf)

# Optional: detailed report
print("\nRandom Forest Classification Report:")
print(classification_report(y_test, y_pred_rf))
```

```python
print("=== Logistic Regression ===")
print("Accuracy :", accuracy)
print("Precision:", precision)
print("Recall   :", recall)
print("F1-score :", f1)

print("\n=== Random Forest ===")
print("Accuracy :", accuracy_rf)
print("Precision:", precision_rf)
print("Recall   :", recall_rf)
print("F1-score :", f1_rf)
```

First we train a simple Logistic Regression model, which is a common baseline for binary classification.  
Then we train a Random Forest model, which uses many decision trees and often finds more complex patterns.


```python
from sklearn.metrics import roc_curve, auc

# Get the predicted probabilities for the positive class (fraud = 1)
# predict_proba returns two columns: prob of class 0, prob of class 1
y_scores_rf = rf_model.predict_proba(X_test)[:, 1]

# Compute ROC curve values
fpr, tpr, thresholds = roc_curve(y_test, y_scores_rf)

# Compute AUC
roc_auc = auc(fpr, tpr)

print("ROC AUC for Random Forest:", roc_auc)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))

# Plot ROC curve
plt.plot(fpr, tpr, label=f"Random Forest (AUC = {roc_auc:.2f})")

# Plot diagonal line for random guessing
plt.plot([0, 1], [0, 1], 'k--', label="Random Guess")

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate (Recall)")
plt.title("ROC Curve - Random Forest")
plt.legend(loc="lower right")

plt.show()
```

We evaluate the models on the test set using accuracy, precision, recall, and F1-score.  
For fraud detection, recall on the fraud class is very important, because it tells us how many frauds we catch.


```python
from sklearn.metrics import precision_recall_curve, average_precision_score

# Compute precision-recall values
precisions, recalls, thresholds_pr = precision_recall_curve(y_test, y_scores_rf)

# Average Precision (area under PR curve)
ap = average_precision_score(y_test, y_scores_rf)

print("Average Precision (AP) for Random Forest:", ap)

plt.figure(figsize=(6, 4))
plt.plot(recalls, precisions, label=f"Random Forest (AP = {ap:.2f})")

plt.xlabel("Recall (how many frauds caught)")
plt.ylabel("Precision (how many flagged are truly fraud)")
plt.title("Precision-Recall Curve - Random Forest")
plt.legend(loc="lower left")
plt.show()
```

We also plot ROC and Precision–Recall curves to see how well the Random Forest separates fraud from normal transactions across different thresholds.
